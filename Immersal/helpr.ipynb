```python
def filter_info_logs(input_file, output_file, key_word=" Info "):
    print(f"Filtering '{key_word.strip()}' logs from {input_file} and saving to {output_file}...")
    with open(input_file, 'r', encoding='utf-8') as infile, open(output_file, 'w', encoding='utf-8') as outfile:
        for line in infile:
            # 1. Only process lines that contain the specified keyword
            if key_word in line:
                # Split the line cleanly at the specified keyword
                parts = line.split(key_word, 1)  # Split into two parts: before and after the keyword
                
                if len(parts) >= 2:
                    before_info = parts[0].strip()  # Contains: "2026/06/11 17:01:09.374 28264 28264"
                    after_info = parts[1].strip()   # Contains: "Unity onResume" or "Unity [WallSolver] ..."
                    
                    # 2. Extract just the time (HH:MM:SS.mmm) from the 'before_info' string
                    # The date is always 10 characters (YYYY/MM/DD), then a space, 
                    # so the time starts at index 11 and ends at 23.
                    time_stamp = before_info[11:23] 
                    
                    # 3. Clean up the message (remove the redundant "Unity " prefix if you want)
                    message = after_info.replace("Unity ", "", 1) if after_info.startswith("Unity ") else after_info
                    
                    # 4. Format and write to file
                    formatted_line = f"{time_stamp} - {message}\n"
                    outfile.write(formatted_line)

# Run the filter
filter_info_logs('_all_logs.txt', '_info_logs.txt', key_word=" Info ")
filter_info_logs('_all_logs.txt', '_error_logs.txt', key_word=" Error ")
print("Filtering complete! Saved to _info_logs.txt and _error_logs.txt")
```
